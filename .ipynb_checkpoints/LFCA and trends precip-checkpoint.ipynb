```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import math
import sys
from sig_proc import lfca
import numpy as np
import scipy as sp
import cartopy
import cartopy.crs as ccrs
import nclcmaps as ncm
from cartopy.util import add_cyclic_point
from datetime import datetime, timedelta
import pandas as pd
```

```python
# import precipitation file
filepath="C:/Users/sydne/OneDrive/Documents/LFCA_precip_temp/LFCA_precip_temp/observations/precip_197901-202212_2.00x2.00deg.nc"
data = xr.open_dataset(filepath)
prc = data['precip']
lat = data['lat']
lon = data['lon']

# calculate anomalies relative to 1981-2010
clim_ref = [datetime(1981,1,1), datetime(2010,12,1)]
dtimes = pd.to_datetime(data['time'].values)
prc_clim = data['precip'].isel(time=np.logical_and(dtimes>=clim_ref[0], dtimes<=clim_ref[-1])).groupby('time.month').mean('time')
prc_anom = data['precip'].groupby('time.month') - prc_clim
      
# partition into a monthly, year, lat, lon array
nyr = (2022-1979 + 1)
nmn = 12
nlt = len(lat)
nln = len(lon)
prc_data = np.zeros((nmn,nyr,nlt,nln))
for mn in range(nmn):
    mon = mn
    for yr in range(nyr):
        prc_data[mn,yr,:,:] = prc_anom[mon,:,:]
        mon = mon+12

# annual mean anomalies 
prc_anom = np.nanmean(prc_data[:,:,:],axis=0)

prc = np.array(prc_anom).transpose(2,1,0)
```

```python
## also temperature file

# import surface temperature file
filepath="C:/Users/sydne/OneDrive/Documents/LFCA_precip_temp/LFCA_precip_temp/observations/temperature_197901-202212_2.00x2.00deg.nc"
data = xr.open_dataset(filepath)
tas_anom =  data['temperature']
tas_clim =  data['climatology']

tas_data = np.zeros((nmn,nyr,nlt,nln))
for mn in range(nmn):
    mon = mn
    for yr in range(nyr):
        tas_data[mn,yr,:,:] = tas_anom[mon,:,:]
        mon = mon+12
        
# calculate anomalies relative to 1981-2010
tas_data = tas_data+np.expand_dims(tas_clim,1)
tas_anom = tas_data - np.expand_dims(np.nanmean(tas_data[:,2:32,:,:],axis=1),1)
        
# annual mean anomalies 
tas_anom = np.nanmean(tas_anom[:,:,:],axis=0)
tas = np.array(tas_anom).transpose(2,1,0)
```

```python
## SEASONAL

prc_anom_jfm = np.nanmean(prc_data[[0,1,2],:,:,:],axis=0)  # jan feb mar
prc_anom_jja = np.nanmean(prc_data[[5,6,7],:,:,:],axis=0)  # june jul aug
prc_anom_djf = np.nanmean(prc_data[[0,1,11],:,:,:],axis=0)
```

```python
## pull spatial linear trend from observations
prc_linear = np.zeros((91,181))
years = range(nyr) + np.ones(nyr)*1979

for lat_ in range(0,91):
    for lon_ in range(0,181):
        prc_anom_pt_ = prc_anom[:,lat_,lon_]  # annual
        m,b,rval,pval,errval = sp.stats.linregress(years, y=prc_anom_pt_, alternative='two-sided')
        prc_linear[lat_,lon_] = m
```

```python
## pull spatial linear trend SEASONAL from obs

prc_linear_jfm = np.zeros((91,181))
prc_linear_jja = np.zeros((91,181))
prc_linear_djf = np.zeros((91,181))

for lat_ in range(0,91):
    for lon_ in range(0,181):
        prc_anom_pt_djf = prc_anom_djf[:,lat_,lon_]
        prc_anom_pt_jja = prc_anom_jja[:,lat_,lon_]
        m_djf,b,rval,pval,errval = sp.stats.linregress(years, y=prc_anom_pt_djf, alternative='two-sided')
        m_jja,b,rval,pval,errval = sp.stats.linregress(years, y=prc_anom_pt_jja, alternative='two-sided')
        prc_linear_djf[lat_,lon_] = m_djf
        prc_linear_jja[lat_,lon_] = m_jja
```

## LFCA

```python
# legacy code to normalize by trace of covariance
variable =  np.transpose(prc_anom_djf,(2,1,0))  #prc     # SEASONAL JFM
s = variable.shape
y, x = np.meshgrid(lat,lon)
area = (np.cos(y*np.pi/180.))
area[np.where(np.isnan(np.nanmean(variable,axis=2)))] = np.nan

domain = np.ones(area.shape)

order = 'C'
x = np.transpose(np.reshape(variable,(s[0]*s[1],s[2]),order=order))
area_weights = np.transpose(np.reshape(area,(s[0]*s[1],1),order=order))
domain = np.transpose(np.reshape(domain,(s[0]*s[1],1),order=order))
icol_ret = np.where((area_weights!=0) & (domain!=0))
icol_disc = np.where((area_weights==0) | (domain==0))
x = x[:,icol_ret[1]]
area_weights = area_weights[:,icol_ret[1]]
normvec = np.transpose(area_weights)/np.sum(area_weights)
scale = np.sqrt(normvec)

isn=np.isnan(x)
x[isn]=0

cov_x = np.cov(x)
variable1_scale =np.trace(cov_x)
```

```python
# input into LFCA
field  = np.transpose(prc_anom_djf,(2,1,0))/variable1_scale   # CHANGE VARIABLE 
lon_n  = lon
```

```python
# more code to get input ready for LFCA
s = field.shape
y, x = np.meshgrid(lat,lon_n)
area = (np.cos(y*np.pi/180.))
area[np.where(np.isnan(np.nanmean(field[:,:,:],axis=2)))] = 0

domain = np.ones(area.shape)

order = 'C'
x = np.transpose(np.reshape(field,(s[0]*s[1],s[2]),order=order))
area_weights = np.transpose(np.reshape(area,(s[0]*s[1],1),order=order))
domain = np.transpose(np.reshape(domain,(s[0]*s[1],1),order=order))
icol_ret = np.where((area_weights!=0) & (domain!=0))
icol_disc = np.where((area_weights==0) | (domain==0))
x = x[:,icol_ret[1]]
area_weights = area_weights[:,icol_ret[1]]
normvec = np.transpose(area_weights)/np.sum(area_weights)
scale = np.sqrt(normvec)

isn=np.isnan(x)
x[isn]=0
```

```python
# two parameters for LFCA
cutoff = 10
truncation = 15

lfcs, lfps, weights, r, pvar, pcs, eofs, ntr, pvar_slow, pvar_lfc, r_eofs, pvar_slow_eofs = lfca(x, cutoff, truncation, scale)
```

```python
nins = np.size(icol_disc[1])
nrows = lfps.shape[0]
lfps_aug = np.zeros((nrows,lfps.shape[1]+nins))
lfps_aug[:] = np.nan
lfps_aug[:,icol_ret[1]] = lfps
nrows = eofs.shape[0]
eofs_aug = np.zeros((nrows,eofs.shape[1]+nins))
eofs_aug[:] = np.nan
eofs_aug[:,icol_ret[1]] = eofs

nlat = len(lat)
nlon = len(lon_n)
lfps_all=np.reshape(lfps_aug,(truncation,nlon,nlat))
lfps_all[np.where(np.abs(lfps_all)>1e5)] = np.nan

lfcs_all = lfcs.T

lfps_f = lfps_all
lfps_f = np.array(lfps_f).transpose(0,2,1)*variable1_scale
lfcs_f = lfcs_all
pvar_f = pvar_slow
r_f    = r
```

```python
# Save data in netCDF file
from netCDF4 import Dataset 
from numpy import dtype
seasonal= 'djf'
ncout = Dataset('netCDF_files/LFCA_GPCP_precipitation_'+str(truncation)+'_Eofs_'+str(cutoff)+'yr_cutoff_03-24-2023'+seasonal+'.nc', 'w', format='NETCDF4')

ncout.createDimension('num' , truncation)
ncout.createDimension('year', nyr)
ncout.createDimension('lat' , nlt)
ncout.createDimension('lon' , nln)

vout1 = ncout.createVariable('lat', dtype('double').char,('lat'))
vout1.long_name = ''
vout1.units = ''

vout2 = ncout.createVariable('lon', dtype('double').char,('lon'))
vout2.long_name = ''
vout2.units = ''

vout3 = ncout.createVariable('lfcs_f', dtype('double').char,('num','year'))
vout3.long_name = ''
vout3.units = ''

vout4 = ncout.createVariable('lfps_f', dtype('double').char,('num','lat','lon'))
vout4.long_name = ''
vout4.units = ''

vout5 = ncout.createVariable('pvar_f', dtype('double').char,('num'))
vout5.long_name = ''
vout5.units = ''

vout6 = ncout.createVariable('r_f', dtype('double').char,('num'))
vout6.long_name = ''
vout6.units = ''

vout1[:] = lat 
vout2[:] = lon 
vout3[:,:] = lfcs_f.real
vout4[:,:,:] = lfps_f[:,:,:].real 
vout5[:] = pvar_f.real
vout6[:] = r_f.real

ncout.close()
```

```python

## pull spatial linear trend from precip observations in 2000-2012 when there was a hiatus in global warming
yr_a = 2000 - 1979
yr_b = 2012 - 1979

prc_linear_0012 = np.zeros((91,181))
years = range(nyr) + np.ones(nyr)*1979

for lat_ in range(0,91):
    for lon_ in range(0,181):
        prc_anom_pt_ = prc_anom[yr_a:yr_b,lat_,lon_]  # annual
        m,b,rval,pval,errval = sp.stats.linregress(years[21:33], y=prc_anom_pt_, alternative='two-sided')
        prc_linear_0012[lat_,lon_] = m
```

```python
cmap = ncm.cmap('MPL_GnBu')
cb_min = -0.2
cb_max = 0.2

fs = 14
f=plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
plt.title('Observed linear precip trend, '+str(years[yr_a].round())+':'+str(years[yr_b].round()))

data = np.array(prc_linear_0012)

datamax = np.max(data)
datamin = np.min(data)

data[data<-1]=-1
data[data>+1]=+1
data_n, lon_n = add_cyclic_point(data, coord=lon)
cf=plt.contourf(lon_n, lat, data_n, np.linspace(cb_min,cb_max,21),cmap=cmap, transform=ccrs.PlateCarree())

cb=plt.colorbar(cf,fraction = 0.023,pad=0.025)
cb.set_label('mm / yr')
cb.set_ticks(np.arange(-cb_min,cb_max,21))
ax.coastlines()

cb.set_ticks(np.linspace(cb_min,cb_max,11))

   
plt.rcParams.update({'font.size':fs})
plt.tight_layout()
```

## Plotting LFCs, LFPs

```python
## PLOT LFCs

for mode in range(10):
    plt.figure(figsize=(6,4))
    var = lfcs[mode,:].real
    from scipy import stats

    time = np.arange(1979,2023,1)

    pos_signal = var.copy()
    neg_signal = var.copy()

    pos_signal[pos_signal <= 0] = np.nan
    neg_signal[neg_signal >= 0] = np.nan

    plt.fill_between(time, var, 0, where=(var >= 0), facecolor='C3', edgecolor='C3', alpha=0.5, interpolate=True,lw=0)
    plt.fill_between(time, var, 0, where=(var <= 0), facecolor='C0', edgecolor='C0', alpha=0.5, interpolate=True,lw=0)
    plt.plot(time,pos_signal,'o',markersize=5,markerfacecolor='k',markeredgecolor='k')
    plt.plot(time,neg_signal,'o',markersize=5,markerfacecolor='k',markeredgecolor='k')
    
    plt.plot(time,var,c='k',lw=1.5)
    
    plt.title('LFC '+ str(mode+1),fontsize=14)
    
    plt.xlabel('Years')
    plt.ylabel('Standard deviation [ % ]')
    plt.ylim(-2.75,2.75)
    plt.xlim(1975,2025)
    plt.rcParams.update({'font.size':14})
    ax=plt.gca()
    plt.setp(ax.spines.values(), linewidth=1.75)
    ax.xaxis.set_tick_params(width=1.75)
    ax.yaxis.set_tick_params(width=1.75)

    plt.tight_layout()
```
